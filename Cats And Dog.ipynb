```python
#training_set
cats = glob.glob('training_set/cats/*.jpg')
dogs = glob.glob('training_set/dogs/*.jpg')
#test_set
cats_test = glob.glob('test_set/cats/*.jpg')
dogs_test = glob.glob('test_set/dogs/*.jpg')
```

```python
print(f'Number of images in Cats is {len(cats)}')
print(f'Number of images in Dogs is {len(dogs)}')
print(f'Number of images in Cats_test is {len(cats_test)}')
print(f'Number of images in Dogs_test is {len(dogs_test)}')
```

```python
#training_set
image = []
label = [] 
for i in cats:
    img = plt.imread(i)
    image.append(img)
    label.append(0)
```

```python
for i in range (20):
    plt.subplot(4,5,i+1)
    plt.imshow(image[i])
plt.tight_layout()
```

```python
for o in dogs:
    img = plt.imread(o)
    image.append(img)
    label.append(1)
```

```python
image[0]
```

```python
label
```

```python
#test_set
image_test = []
label_test = [] 
for i in cats_test:
    img = plt.imread(i)
    image.append(img)
    label.append(0)
for i in dogs_test:
    img = plt.imread(i)
    image_test.append(img)
    label_test.append(1)
```

```python
image_test,label_test
```

### Shuffle the tarining and testing set

```python
#training_set
image,label = shuffle(image,label,random_state=100)
```

```python
#testing_set
image_test,label_test = shuffle(image_test,label_test)
```

```python
#Plot this
for i in range (20):
    plt.subplot(4,5,i+1)
    plt.imshow(image[i])
    if label[i]==1:
        plt.title('Dogs')
    else:
        plt.title('Cats')
plt.tight_layout()
```

### Resize traning and testing set 


```python
# training_set
xtrain = []
for i in image:
    xtrain.append(resize(i,(64,64,3)))
# make it in array format
xtrain = np.array(xtrain)
```

```python
xtrain
```

```python
# training_set
xtest = []
for i in image_test:
    xtest.append(resize(i,(64,64,3)))
# make it in array format
xtest= np.array(xtest)
```

```python
xtest
```

```python
#Plot this
for i in range (20):
    plt.subplot(4,5,i+1)
    plt.imshow(xtrain[i])
    if label[i]==1:
        plt.title('Dog')
    else:
        plt.title('Cat')
plt.tight_layout()
```

```python
ytrain = np.array(label)
ytest = np.array(label_test)
```

```python
ytrain,ytest
```

### Create CNN Model

```python
model = keras.models.Sequential([
# first layer
    keras.layers.Conv2D(16,(3,3),activation='relu',input_shape=(64,64,3)),
    keras.layers.MaxPool2D(2,2),
# second layer 
    keras.layers.Conv2D(32,(3,3),activation='relu'),
    keras.layers.MaxPool2D(2,2),
# Faltten Layer 
    keras.layers.Flatten(),
# Dense Layer 
    keras.layers.Dense(64,activation='relu'),
#output layer
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
# Compile the model
model.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
# train the model 
model.fit(xtrain,ytrain,epochs=10,validation_data = [xtest,ytest])
```

```python
# evaluate the model
model.evaluate(xtrain,ytrain)
```

```python
# Predict the  model
Prediction = model.predict(xtest)
```

```python
Prediction = list(map(lambda x:1 if x>0.5 else 0,Prediction))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_predictions(ytest,Prediction)
```
